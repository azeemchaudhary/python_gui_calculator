```python
from tkinter import *
import parser as ps
```

```python
root = Tk()
root.title("Calculator")

#Calculate
def calculate():
    get_input = cal_screen.get()
    try:
        a = ps.expr(get_input).compile()
        result = eval(a)
        clear_all()
        cal_screen.insert(0, result)
    except Exception:
        clear_all()
        #cal_screen.insert(0, "Error")
        
#calculate factorial
def get_fact():
    n = int(cal_screen.get())
    fact = 1
    if n < 0:
        clear_all()
        cal_screen.insert(0, 0)
    elif n == 0 or n == 1:
        clear_all()
        cal_screen.insert(0, 1)
    else:
        for i in range(1, n + 1):
            fact = fact * i
            clear_all()
        cal_screen.insert(0, fact)
        
        
#get button values
i = 0
def get_value(num):
    global i
    cal_screen.insert(i, num)
    i+=1
    
#clear all screen
def clear_all():
    cal_screen.delete(0, END)
    
#clear one by one
def undo():
    get_input = cal_screen.get()
    if len(get_input) > 0:
        new_input = get_input[:-1]
        clear_all()
        cal_screen.insert(0, new_input)
        
#get operators
def get_operator(opt):
    global i
    opt_len = len(opt)
    cal_screen.insert(i, opt)
    i = i + opt_len
    

    
cal_screen = Entry(root)
cal_screen.grid(row = 0, columnspan = 6, sticky = W + E)

Button(root, text = "1", command = lambda :get_value(1)).grid(row = 1, column = 0, padx = 3, pady = 3)
Button(root, text = "2", command = lambda :get_value(2)).grid(row = 1, column = 1, padx = 3, pady = 3)
Button(root, text = "3", command = lambda :get_value(3)).grid(row = 1, column = 2, padx = 3, pady = 3)

Button(root, text = "4", command = lambda :get_value(4)).grid(row = 2, column = 0, padx = 3, pady = 3)
Button(root, text = "5", command = lambda :get_value(5)).grid(row = 2, column = 1, padx = 3, pady = 3)
Button(root, text = "6", command = lambda :get_value(6)).grid(row = 2, column = 2, padx = 3, pady = 3)

Button(root, text = "7", command = lambda :get_value(7)).grid(row = 3, column = 0, padx = 3, pady = 3)
Button(root, text = "8", command = lambda :get_value(8)).grid(row = 3, column = 1, padx = 3, pady = 3)
Button(root, text = "9", command = lambda :get_value(9)).grid(row = 3, column = 2, padx = 3, pady = 3)

Button(root, text = "AC", command = lambda :clear_all()).grid(row = 4, column = 0, padx = 3, pady = 3)
Button(root, text = "0", command = lambda :get_value(0)).grid(row = 4, column = 1, padx = 3, pady = 3)
Button(root, text = "=", command = lambda :calculate()).grid(row = 4, column = 2, padx = 3, pady = 3)

Button(root, text = "*", command = lambda :get_operator("*")).grid(row = 1, column = 3, padx = 3, pady = 3)
Button(root, text = "/", command = lambda :get_operator("/")).grid(row = 2, column = 3, padx = 3, pady = 3)
Button(root, text = "-", command = lambda :get_operator("-")).grid(row = 3, column = 3, padx = 3, pady = 3)
Button(root, text = "+", command = lambda :get_operator("+")).grid(row = 4, column = 3, padx = 3, pady = 3)

Button(root, text = "<--", command = lambda :undo()).grid(row = 1, column = 4, padx = 3, pady = 3)
Button(root, text = "(", command = lambda :get_operator("(")).grid(row = 2, column = 4, padx = 3, pady = 3)
Button(root, text = "%", command = lambda :get_operator("%")).grid(row = 3, column = 4, padx = 3, pady = 3)
Button(root, text = "exp", command = lambda :get_operator("**")).grid(row = 4, column = 4, padx = 3, pady = 3)

Button(root, text = "pi", command = lambda :get_operator("*3.14")).grid(row = 1, column = 5, padx = 3, pady = 3)
Button(root, text = ")", command = lambda :get_operator(")")).grid(row = 2, column = 5, padx = 3, pady = 3)
Button(root, text = "x!", command = lambda :get_fact()).grid(row = 3, column = 5, padx = 3, pady = 3)
Button(root, text = "^2", command = lambda :get_operator("**2")).grid(row = 4, column = 5, padx = 3, pady = 3)

root.mainloop()
```
